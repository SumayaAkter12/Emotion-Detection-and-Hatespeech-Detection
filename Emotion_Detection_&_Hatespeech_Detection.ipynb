
#Load Packages

```python
from pysentimiento import create_analyzer
```

#Emotion Detection


```python
emotion = create_analyzer(task = 'emotion',lang='en')
result = emotion.predict("i am happy")
result.output
```

```python
def emotion_dection(text):
    emotion = create_analyzer(task = 'emotion',lang='en')
    result = emotion.predict(text)
    return result.output

emotion_dection("I am happy")
```

#hate speech detection





```python
hatespeech = create_analyzer(task = 'hate_speech',lang='en')
result = hatespeech.predict("I hate eating apples , that's why i don't want to sit with apple seller")
next(iter(result.probas.items()),None)
```

```python
def hatespeech_detection(text):
    hatespeech = create_analyzer(task = 'hate_speech',lang='en')
    result = hatespeech.predict(text)
    return next(iter(result.probas.items()),None)
    hatespeech_detection("I hate eating apples , that's why i don't want to sit with apple seller")

```

#Prediction system





```python
text = """ I woke up this morning feeling incredibly happy and grateful for the beautiful weather outside"""
```

```python
emo = emotion_dection(text)
hate = hatespeech_detection(text)
print ("Emotion :",emo)
print ("Speech  :",hate)
```
